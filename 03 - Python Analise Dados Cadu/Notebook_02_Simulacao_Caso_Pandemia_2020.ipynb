```python
import numpy as np
import plotly.express as px
```

```python
x = 1100 # numero de infectados
h = np.random.binomial(x,0.15,10000) # sorteando no. de hospitalizados 

(h > 200).mean()  # percentual de vezes que houve leito 
```

```python
X = np.random.binomial(1300,0.15,100000)
```

```python
X = range(1000,1400,5) # populao infectada ao mesmo tempo

Y = np.array([])

for x in X:
    h = np.random.binomial(x,0.15,20000) # sorteando hospitalizacao
    
    Y = np.append(Y,(h > 200).mean()) # percentual de leitos excedentes
```

```python
px.scatter(x=X,y=Y)
```

Model completo 

```python
X = np.random.binomial(100000,0.025,10000)

Y = np.array([])
for x in X:
    h = np.random.binomial(x,0.15,1)
    Y = np.append(Y,h)
    
Z = np.array([])
for y in Y:
    ob = np.random.binomial(y,0.3,1)
    Z = np.append(Z,ob)
```
