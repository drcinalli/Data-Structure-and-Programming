**Exercício de simulação**

A urna I contém uma bola vermelha e duas brancas. A urna II contém duas bolas vermelhas e uma branca. Tiramos aleatoriamente um bola da urna I, colocamos na urna II e misturamos. Em seguida, tiramos aleatoriamente uma bola da urna II. Qual é a probabilidade de tiramos uma bola branca da urna II?

*Solução analítica*
0 - bola vermelha
1 - bola branca

Pr(bola_urna_I = 1) = 2/3
Pr(bola_urna_I = 0) = 1/3

Pr(bola_urna_II = 1 | bola_urna_I) = 
      Pr(bola = 1 | bola_urna_I = 1) * Pr(bola_urna_I = 1) +
      Pr(bola = 1 | bola_urna_I = 0) * Pr(bola_urna_I = 0)

Pr(bola = 1 | bola_urna_I = 1) = 1/2
Pr(bola = 1 | bola_urna_I = 0) = 1/4

Pr(bola_urna_II = 1 | bola_urna_I) = 2/3 * 1/2 + 1/3 * 1/4 = 5/12 = 0.41666..


**Simulação**

Simulação I - Usando apenas distribuição uniforme.

```
U_1 = np.array([1,1,0]) # urna I
U_2 = np.array([1,0,0]) # urna II

res = np.array([]) # vetor para armazenar resultado
n = 200000 # numero de simulacoes a realizar

for i in range(0,n):
    
    bola = U_1[np.random.randint(0,3,1)] # selecao aletoria de uma bola da urna I
    
    U_2_n = np.append(U_2,bola) # adicionando a bola retirada da urna I na urna II
    
    bola = U_2_n[np.random.randint(0,4,1)] # selecao aletoria de uma bola da urna II 

    res = np.append(res, bola)
    
    
print(res.mean())
```

Simualação II - Usando distribuição binomial.

```
U_1 = np.array([1,1,0]) # urna I
U_2 = np.array([1,0,0]) # urna II

res = np.array([]) # vetor para armazenar resultado
n = 200000 # numero de simulacoes a realizar

bolas = np.random.binomial(1,2/3,n) # selecao aletoria de uma bola da urna I
    
bolas_brancas_urna_II = bolas + 1 # numero de bolas brancas na urna II    

pr_bolas_brancas_urna_II = bolas_brancas_urna_II/4 # probabilidade de bolas brancas na urna II    
       
print(pr_bolas_brancas_urna_II.mean())
```
