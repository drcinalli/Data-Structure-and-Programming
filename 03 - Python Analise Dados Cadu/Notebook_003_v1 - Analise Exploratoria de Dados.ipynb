<H1>Análise exploratória de dados</H1>

O objetivo deste notebook é oferecer um guia prático básico de como conduzir uma **análise exploratória de dados**.

<h2>Introdução</h2>
    
A **análise exploratória de dados – AED** – é uma das estratégias mais conhecidas de análise de dados. O primeiro trabalho sobre AED foi escrito por John Tukey ainda na década de 60. Via de regra, todo trabalho de **analytics** deve começar com uma boa AED. É através da AED que conseguirmos desenvolver *insights* a respeito das características dos dados e eventualmente especular ou verificar determinadas características esperadas nos dados.




<h3>Maninupulação de dados</h3>

Antes de começarmos a falar efetivamente de AED, precisamos de uma boa ferramenta de manipulação de dados, que nos permita realizar as consultas e seleções que desejamos em nossas análises e investigações.

**PANDAS**

O pandas é um *framework* de manipulação de dados disponível em python. Veja o site: https://pandas.pydata.org/. Com o pandas podemos fazer buscas, consultas e manipulações de um modo geral com rapidez e eficiência, mesmo com volumes de dados considerados grandes.
